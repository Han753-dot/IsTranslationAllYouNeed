```python
import json
from ollama import chat
from ollama import ChatResponse
```

```python
json_file_path16 = "..\\data\\datasets_json\\rest16.json"
json_file_path15 = "..\\data\\datasets_json\\rest15.json"
```

```python
# Read the XML file
with open(json_file_path15) as json_file:
    rest15_dict = json.load(json_file)
```

```python
def translate(json_dict: dict) -> dict:
    """Translates review texts with Google Translate"""
    
    for idx, review in enumerate(json_dict, start=1):
         
        review_text = review['text']


        #Translation here
        response: ChatResponse = chat(model='llama3.2', messages=[
            {
        'role': 'user',
        'content': f"Translate the following text from English into German. Provide only the translated string, without any additional comments, annotations or symbols: {review_text}. ",
            },
        ])

        translated = response.message.content

        print(f"Translated Sentence {idx}: {translated}")
        
        review['text'] = translated

    return json_dict
```

```python
dict_translated = translate(rest15_dict)
```

```python
with open("rest15_acsa_llama_translated.json", "w") as outfile:
    json.dump(dict_translated, outfile)
```
